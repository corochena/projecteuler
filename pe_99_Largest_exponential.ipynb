```python
import urllib.request as url
import math

f = url.urlopen('https://projecteuler.net/project/resources/p099_base_exp.txt')

maxlog = 0
i = 1

for line in f:
  txt = line.decode('utf8').split(',')
  base = int(txt[0])
  exp = int(txt[1][0:-1])
  x = exp * math.log(base)
  if x > maxlog:
    maxlog = x
    maxbase = base
    maxexp = exp
    index = i
  i += 1

print(index, maxlog, maxbase, maxexp)
```
