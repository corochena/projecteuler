```python
def sqrconv():
  n = 0
  d = 1
  count = 0
  for i in range(1, 1001):
    temp = n
    n = d
    d = 2*d + temp
    if len(str(n+d)) > len(str(d)):
      count += 1
  print(count)

sqrconv()
```
