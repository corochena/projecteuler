```python
def primesOrig(n):
  primes = [2]

  for num in range(primes[-1] + 1, n + 1):
    isPrime = True
    for div in primes:
      if num % div == 0:
        isPrime = False
        # print(num, ' is divisible by ', div, ' try next prime divisor')
        break
      if div > num**0.5:
        # print(div, ' > raiz de ', num, ' enough, try next prime divisor')
        break
    if isPrime:
      # print('prime found ', num)
      primes.append(num)

  return primes
```

```python
primelist = primesOrig(9999)
```

```python
minuslist = primesOrig(1000)
```

```python
for p in minuslist:
  primelist.remove(p)
```

```python
# using SETS
def permstr(s):
  if len(s) == 1:
    return {s}
  else:
    elems = sorted([e for e in s])
    combo = set()
    for e in elems:
      next = elems[:]
      next.remove(e)
      combo.update(prefixset(e, permstr(''.join(next))))
    return combo

def prefixset(pf, elems):
  return {pf + elem for elem in elems}
```

```python
candidates = []

for p in primelist:
  permutationset = permstr(str(p))
  #print(permutationset)
  numprime = 0
  plist = []
  for x in permutationset:
    if int(x) in primelist:
      plist.append(int(x))
      numprime += 1
  if numprime >= 3:
    #print(permutationset)
    if plist not in candidates:
      candidates.append(plist)
print(candidates)
```

```python
def checkdiff(l):
  l = sorted(l)
  n = len(l)
  for i in range(n-2):
    first = l[i]
    for j in range(i+1,n-1):
      second = l[j]
      for k in range(j+1, n):
        third = l[k]
        if third - second == second - first:
          return (first, second, third)
  return False
```

```python
for l in candidates:
  answer = checkdiff(l)
  if answer:
    print(answer)
    print(l)
```
