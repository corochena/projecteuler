```python
def primesOrig(n):
  primes = [2]

  for num in range(primes[-1] + 1, n + 1):
    isPrime = True
    for div in primes:
      if num % div == 0:
        isPrime = False
        # print(num, ' is divisible by ', div, ' try next prime divisor')
        break
      if div > num**0.5:
        # print(div, ' > raiz de ', num, ' enough, try next prime divisor')
        break
    if isPrime:
      # print('prime found ', num)
      primes.append(num)

  return primes

shortlist = primesOrig(10000000)
shortlist.reverse()
```

```python
for i in shortlist:
  stri = str(i)
  if '1' in stri and '2' in stri and '3' in stri and '4' in stri and '5' in stri and '6' in stri and '7' in stri:
    print(i)
    break
```
