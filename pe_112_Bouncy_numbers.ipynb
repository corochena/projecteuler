```python
# Problem 112: Bouncy numbers

bouncy = 0
n = 99

while 100*bouncy != n * 99:
    inc = False
    dec = False
    n += 1
    s = str(n)
    for k in range(len(s)-1):
        if int(s[k]) > int(s[k+1]):
            dec = True
        elif int(s[k]) < int(s[k+1]):
            inc = True
        if dec and inc:
            bouncy += 1
            break
print(bouncy, n, bouncy/n)
```
