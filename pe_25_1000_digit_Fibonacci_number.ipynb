```python
def fibLen(n):
  a = 1
  b = 1
  i = 2
  ndig = 1
  while ndig != n:
    temp = b
    b = a + b
    a = temp
    ndig = len(str(b))
    i += 1
  print(i, b)

fibLen(1000)
```
