```python
def digitFractions():
  for den in range(11, 100):
    ds = str(den)
    for num in range(10, den):
      ns = str(num)
      frac = num / den
      if ds[1] != '0':
        #print(ds[1], num, den, '1st if')
        if ds[0] == ns[0]:
          if frac == int(ns[1]) / int(ds[1]):
            print(ns[1], "/", ds[1])
        if ds[0] == ns[1]:
          if frac == int(ns[0]) / int(ds[1]):
            print(ns[0], "/", ds[1])
      if ds[0] != '0':
        #print(ds[0], num, den, '2nd if')
        if ds[1] == ns[0]:
          if frac == int(ns[1]) / int(ds[0]):
            print(ns[1], "/", ds[0])
        if ds[1] == ns[1] and ds[1] != '0':
          if frac == int(ns[0]) / int(ds[0]):
            print(ns[0], "/", ds[0])                                
digitFractions()
```
