```python
def palinum(s):
  # Returns True if string s is palindrome
  for i, d in enumerate(s):
    if d != s[-(i+1)]:
      return False
  return True
  
def palin(n):
  paliset = set()
  for i in range(n+1):
    if palinum(str(i)) and palinum(str(bin(i))[2:]):
      paliset.add(i)
  print(sum(paliset))

palin(1000000)
```
