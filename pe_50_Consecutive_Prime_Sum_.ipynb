```python
def primesOrig(n):
  primes = [2]

  for num in range(primes[-1] + 1, n + 1):
    isPrime = True
    for div in primes:
      if num % div == 0:
        isPrime = False
        # print(num, ' is divisible by ', div, ' try next prime divisor')
        break
      if div > num**0.5:
        # print(div, ' > raiz de ', num, ' enough, try next prime divisor')
        break
    if isPrime:
      # print('prime found ', num)
      primes.append(num)

  return primes

primesm = primesOrig(1000000)
print(sum(primesm))
```

```python
for i in range(546):
  a = sum(primesm[0+i:547])
  b = sum(primesm[0:547-i])
  if a in primesm:
    print('a',i,a)
  if b in primesm:
    print('b',i,b)
  if a in primesm or b in primesm:
    break
```
