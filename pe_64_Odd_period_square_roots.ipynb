```python
def continuedFraction(n):
    sqrtn = n**0.5
    if sqrtn == int(sqrtn):
        return int(sqrtn)
    else:
        entero = 0
        numerador = 1
        denominador = 1
        sustraendo = entero - denominador*int(sqrtn)
        ini = sustraendo
        flag = True
        period = 0
        cFracList = [int(sqrtn)]
        while flag:
            period += 1
            denominador = n - sustraendo**2
            if denominador % numerador == 0:
                denominador = int(denominador / numerador)
            entero = int((sqrtn - sustraendo) / denominador)
            cFracList.append(entero)
            sustraendo = -sustraendo - denominador*entero
            numerador = denominador
            if sustraendo == ini and numerador == 1:
                flag = False
    return cFracList

def decimalApprox(contFrac, iter):
    parteEntera = contFrac[0]
    pList = contFrac[1:]
    plen = len(pList)
    if iter > plen:
        n = int(iter / plen)
        more = iter % plen
        pList = pList * n + pList[:more]
    else:
        pList = pList[:iter]
    
    pList.reverse()
    
    frac = 0
    for p in pList:
        frac = 1 / (p + frac)
    return parteEntera + frac

x = 13
exact = x**0.5
cf = continuedFraction(x)
print(cf)

for i in range(1,15):
    approx = decimalApprox(cf, i)
    error = (exact - approx) / exact * 100
    print('iter>%d - sqrt(%d) is exactly %.10f, with approximation %.10f, error = %.3f %%' % (i, x, exact, approx, error))

limit = 10000
oddPeriod = 0

for n in range(2, limit+1):
    contRep = continuedFraction(n)
    if type(contRep) != type(1):
        period = len(contRep) - 1
        if period % 2 != 0:
            oddPeriod += 1
        print('sqrt(%d) = %s, period = %d' % (n, contRep, period))

print('Exactly %d continued fractions have an odd period' % (oddPeriod))
```
