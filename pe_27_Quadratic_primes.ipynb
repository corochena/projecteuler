```python
def primesOrig(n):
  primes = [2]

  for num in range(primes[-1] + 1, n + 1):
    isPrime = True
    for div in primes:
      if num % div == 0:
        isPrime = False
        # print(num, ' is divisible by ', div, ' try next prime divisor')
        break
      if div > num**0.5:
        # print(div, ' > raiz de ', num, ' enough, try next prime divisor')
        break
    if isPrime:
      # print('prime found ', num)
      primes.append(num)

  return primes

shortlist = primesOrig(1000000)
shortset = set(shortlist)
```

```python
maxn = 0

for a in range(-999, 1000):
  for b in range(-999, 1000):
    n = 0
    x = b
    while x in shortset:
      n += 1
      x = n**2 + a*n + b
    if n > maxn:
      maxn = n
      print(a, b, maxn, a*b) 
```
