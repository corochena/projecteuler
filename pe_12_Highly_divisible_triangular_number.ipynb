```python
# esta celda si funciona

import time as t
 
def setDivisors(n):
  # returns a set with divisors of n
  divset = set([1, n])
  for i in range(2, int(n/2)+1):
    if n % i == 0:
      divset.add(i)
  return divset
 
def setDivisors2(n):
  divset = set([1])
  i = 2
  while n != 1:
    if n % i == 0:
      divls = list(divset)
      for elem in divls:
        divset.add(i*elem)
      divset.add(i)
      n /= i
    else:
      i += 1  
  return divset
 
def findHDTN(n):
  k = 7
  divnum = 6
  currTime = t.time()
 
  while divnum < n:
    k += 1
    triangN = int(k*(k+1)/2)
    divnum = len(setDivisors2(triangN))
 
  print(t.time() - currTime)
  print(divnum, triangN, k)
 
findHDTN(500)
#setDivisors(28)
#setDivisors2(76576500)
```
