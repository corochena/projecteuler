```python
# Truncatable primes
# Problem 37

def primesOrig(n):
  primes = [2]

  for num in range(primes[-1] + 1, n + 1):
    isPrime = True
    for div in primes:
      if num % div == 0:
        isPrime = False
        # print(num, ' is divisible by ', div, ' try next prime divisor')
        break
      if div > num**0.5:
        # print(div, ' > raiz de ', num, ' enough, try next prime divisor')
        break
    if isPrime:
      # print('prime found ', num)
      primes.append(num)

  return primes

shortlist = primesOrig(1000000)
shortlist.reverse()
shortset = set(shortlist)
```

```python
trunccount = 0
trunclist = []

for d in shortlist:
  dstr = str(d)
  trunc = True
  for i in range(1,7):
    #print(dstr[:i], dstr[-i:])
    if not int(dstr[:i]) in shortset or not int(dstr[-i:]) in shortset:
      trunc = False
      break
  if trunc:
    trunccount += 1
    trunclist.append(d)
  if trunccount == 17:
    break

print(sum(trunclist) - sum([7,5,3,2]))
```
