```python
def sumDigits(s):
  sum = 0
  for d in s:
    sum += int(d)
  return sum

def findMax():
  n = 100
  maxsum = 0

  for a in range(n):
    for b in range(n):
      s = sumDigits(str(a**b))
      if s > maxsum:
        base = a
        exp = b
        maxsum = s
  print(maxsum, base, exp)

findMax()
```
