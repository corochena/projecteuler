```python
def division(divisor):
  largo = 1
  periodo = "1"
  dividendo = 1000 - divisor
  dividendos = set([dividendo])
  while dividendo != 1:
    periodo += str(int(10*dividendo / divisor))
    dividendo = 10*dividendo % divisor
    if dividendo in dividendos:
      return largo + 1, periodo
    dividendos.add(dividendo)
    largo += 1
  return largo, periodo

bigperiod = 0
for i in range(501, 1000):
  largo, periodo = division(i)
  if largo > bigperiod:
    bigperiod = largo
    print(i, largo, periodo)
```
