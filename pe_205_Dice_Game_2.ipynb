La forma en que lo hice me deja pensando como hacerlo usando funciones para generar la distribucion de probabilidad de cada caso

```python
def distDict(numDice, maxNum):
  # returns a dictionary with probabilities
  dist = {}
  n = maxNum ** numDice
  results = [dectobase(i, maxNum) for i in range(n)]
  results = list(map(lambda s: sum([int(c) for c in s]) + numDice, results))
  rlen = len(results)

  for k in range(numDice, numDice * maxNum + 1):
    dist[k] = results.count(k) / rlen

  return dist  
```

```python
def dectobase(numdec, base):
  s = ''
  dividendo = numdec
  while True:
    s = str(dividendo % base) + s
    dividendo = int(dividendo/base)
    if dividendo == 0:
      return s
```

```python
def basetodec(num, base):
  numstr = str(num)
  total = 0
  for i, c in enumerate(reversed(numstr)):
    total += int(c) * base**i
  return total
```

```python
def main():
  dictPeter = distDict(9, 4)
  dictColin = distDict(6, 6)
  probability = 0

  for kp in dictPeter:
    pcolin = 0
    for kc in dictColin:
      if kc < kp:
        pcolin += dictColin[kc]
    # print(kp, peterdict[kp], pcolin)
    probability += dictPeter[kp]*pcolin

  print(probability)

main()
```
