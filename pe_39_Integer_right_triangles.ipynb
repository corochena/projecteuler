```python
def sides(p):
  solutions = set()
  endx = int(p/(2 + 2**0.5))
  for x in range(1, endx):
    starty = int((p-x)/2)
    for y in range(starty, 0, -1):
      z = (x**2 + y**2)**0.5
      if x + y + z == p:
        solutions.add((x, y, z))
  return solutions

maxsolutions = 0
for p in range(1001):
  nsol = len(sides(p))
  if nsol > maxsolutions:
    maxsolutions = nsol
    print(p, maxsolutions)

print(maxsolutions)
```
