```python
def sumPow(n):
  sum = 0
  for i in range(1, n+1):
    sum += i**i
  return sum

print(str(sumPow(1000))[-10:])
```

```python
# una forma abreviada usando una list comprehension
str(sum([x**x for x in range(1, 1001)]))[-10:]
```
