```python
def samedig(n1, n2):
  n1set = set([(d, n1.count(d)) for d in n1])
  n2set = set([(d, n2.count(d)) for d in n2])
  return n1set == n2set

def permmult(lim):
  for i in range(11,lim):
    test = [str(n*i) for n in (1,2,3,4,5,6)]
    if samedig(test[0], test[1]):
      if samedig(test[0], test[2]):
        if samedig(test[0], test[3]):
          if samedig(test[0], test[4]):
            if samedig(test[0], test[5]):
              return test

permmult(1000000)
```
