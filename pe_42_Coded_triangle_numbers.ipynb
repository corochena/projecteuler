```python
import urllib.request as url

infile =url.urlopen('https://projecteuler.net/project/resources/p042_words.txt')

for line in infile:
  words = line.decode('utf8').split(',')

words = list(map(lambda word: word[1:-1], words))

alpha = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

triangset = set(map(lambda k: int(k*(k+1)/2), range(1,101)))

def nameVal(name):
  val = 0
  for letter in name:
    val += alpha.index(letter) + 1
  return val

triangwords = 0

for word in words:
  if nameVal(word) in triangset:
    triangwords += 1

print(triangwords)
```
