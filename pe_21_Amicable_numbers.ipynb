```python
def properDivisors(n):
  divset = set([1])
  i = 2
  while n != 1:
    if n % i == 0:
      divls = list(divset)
      for elem in divls:
        divset.add(i*elem)
      divset.add(i)
      n /= i
    else:
      i += 1  
  n = max(divset)
  divset.remove(n)
  return divset

def amicables(a, b):
  return sum(properDivisors(a)) == b and sum(properDivisors(b)) == a

def amicablesUnder(n):
  totalamicables = 0
  for a in range(4, n):
    adiv = properDivisors(a)
    for b in range(a + 1, n):
      if sum(adiv) == b:
        if sum(properDivisors(b)) == a:
          print(a, b)
          totalamicables += a + b
  print(totalamicables)

amicablesUnder(10000)
```
