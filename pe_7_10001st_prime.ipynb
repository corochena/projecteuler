```python
def nPrimes(n):
  # retuns n primes
  primes = set([2])
  size = 1
  i = 3

  while size < n:
    isPrime = True
    for p in primes:
      if i % p == 0:
        isPrime = False
        i += 1
        break
    if isPrime:
      primes.add(i)
      last = i
      i += 1
      size += 1
  
  return last, primes

print(nPrimes(10001))
```
