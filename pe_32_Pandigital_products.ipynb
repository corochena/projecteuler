```python
products = set()
 
for a in range(1, 10000):
  for b in range(1, 2000):
    prod = a*b
    s = str(a) + str(b) + str(prod)
    if len(s) == 9 and '1' in s and '2' in s and '3' in s and '4' in s and '5' in s and '6' in s and '7' in s and '8' in s and '9' in s:
      print(a, b, prod)
      products.add(prod)
 
print(sum(products))
print(products)
```

```python
products = set()
 
for a in range(1, 10000):
  for b in range(1, 2000):
    prod = a*b
    s = str(a) + str(b) + str(prod)
    if len(s) == 9:
      flag = True
      for digit in '123456789':
        if digit not in s:
          flag = False
          break
      if flag:
        print(a, b, prod)
        products.add(prod)
 
print(sum(products))
print(products)
```
