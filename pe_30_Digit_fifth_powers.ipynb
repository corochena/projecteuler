```python
def fifthpowers():
  maxnum = 6*9**5
  sumfifth = 0
  for i in range(11,maxnum+1):
    strnum = str(i)
    temptotal = 0
    for digit in strnum:
      temptotal += int(digit) ** 5
    if temptotal == i:
      print(temptotal)
      sumfifth += temptotal
  return sumfifth

fifthpowers()
```
