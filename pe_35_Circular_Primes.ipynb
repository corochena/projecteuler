```python
def primesOrig(n):
  primes = [2]

  for num in range(primes[-1] + 1, n + 1):
    isPrime = True
    for div in primes:
      if num % div == 0:
        isPrime = False
        # print(num, ' is divisible by ', div, ' try next prime divisor')
        break
      if div > num**0.5:
        # print(div, ' > raiz de ', num, ' enough, try next prime divisor')
        break
    if isPrime:
      # print('prime found ', num)
      primes.append(num)

  return primes

plist = primesOrig(1000000)
print(len(plist))
```

```python
pset = set(primesOrig(1000000))
print(len(pset))
```

```python
def rotations2(num):
  rset = set()
  s = str(num)
  for i in range(1, len(s)):
    rot = int(s[i:] + s[:i])
    rset.add(rot)
  return rset
```

```python
def rotations(num):
  rlist = []
  s = str(num)
  for i in range(1, len(s)):
    rot = int(s[i:] + s[:i])
    rlist.append(rot)
  return rlist
```
