```python
def collatz(n):
  elems = []
  while n != 1:
    if n % 2 == 0:
      elems.append(int(n/2))
      n = n/2
    else:
      elems.append(int(3*n + 1))
      n = 3*n + 1
  return len(elems), elems

def findLongest():
  longest = 0
  for i in range(2,1000000):
    size, seq = collatz(i)
    if size > longest:
      longest = size
      x = i
      longseq = seq
  print(x, longseq)

findLongest()
```
