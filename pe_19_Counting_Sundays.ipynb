```python
import datetime as dt

fecha1 = dt.date(1901,1,1)
fecha2 = dt.date(2000, 12, 31)
undia = dt.timedelta(days = 1)

fecha = fecha1
domingos = 0

while fecha <= fecha2:
  if fecha.day == 1 and fecha.weekday() == 6:
    domingos += 1
  fecha = fecha + undia

print(domingos)
```
