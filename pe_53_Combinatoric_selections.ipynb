```python
import math

gtonemillion = 0

for n in range(1, 101):
  nfact = math.factorial(n)
  for r in range(1, n + 1):
    if nfact / math.factorial(r) / math.factorial(n - r) > 1e6:
      gtonemillion += 1
      #print(n, r)
print(gtonemillion)
```
