```python
def trianpenhex(start, nlim):
  trilist = [int(n*(n+1)/2) for n in range(start,nlim)]
  penset = set([int(n*(3*n-1)/2) for n in range(2,nlim)])
  hexset = set([int(n*(2*n-1)) for n in range(2,nlim)])
  
  for num in trilist:
    if num in penset:
      if num in hexset:
        #print(num, num in penset)
        return num, trilist.index(num)

trianpenhex(286,80000)
```
