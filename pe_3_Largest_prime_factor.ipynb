```python
def primesOrig(n):
  # esta funcion devuelve una lista de nmeros primos menores que n
  primes = [2]

  for num in range(primes[-1] + 1, n + 1):
    isPrime = True
    for div in primes:
      if num % div == 0:
        isPrime = False
        # print(num, ' is divisible by ', div, ' try next prime divisor')
        break
      if div > num**0.5:
        # print(div, ' > raiz de ', num, ' enough, try next prime divisor')
        break
    if isPrime:
      # print('prime found ', num)
      primes.append(num)

  return primes

def primeDivisors(n):
  primesList = primesOrig(100000)
  divisors = set()
  for p in primesList:
    if p > n:
      break
    if n % p == 0:
      divisors.add(p)
      n = n/p
  return max(divisors), divisors

print(primeDivisors(600851475143))
```
