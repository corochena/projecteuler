```python
def sumsqdig(n):
  sn = str(n)
  sum = 0
  for d in sn:
    sum += int(d)**2
  if sum == 89:
    return True
  elif sum == 1:
    return False
  else:
    return sumsqdig(sum)

total = 0
for i in range(1,10000000):
  if sumsqdig(i):
    total += 1
print(total)
```
