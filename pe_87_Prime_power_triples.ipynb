Basicamente tuve que sacar la raiz cuadrada de 50,000,000 para saber hasta donde tenia que chequear en la lista de primos

```python
def primesUpto(n):
  integers = set(range(2, n+1))
  compounds = set()

  for i in integers:
    k = 2
    while k*i <= n:
      compounds.add(k*i)
      k += 1
    if i > n**0.5:
      break
  return integers.difference(compounds)
```

```python
primes = primesUpto(7072)
```

```python
primes = sorted(primes)
```

```python
nums = set()
lim = 50000000

for x in primes:
  if x**4 >= lim:
    print(x,y,z,'break x^4')
    break
  for y in primes:
    if x**4 + y**3 >= lim:
      print(x,y,z,'break x^3')
      break
    for z in primes:
      n = x**4 + y**3 + z**2
      if n < lim:
        nums.add(n)
      else:
        print(x,y,z,'print break x^2')
        break
```
