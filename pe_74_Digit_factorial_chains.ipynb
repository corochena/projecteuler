```python
import math

def sumfactdig(n):
  s = str(n)
  total = 0
  for d in s:
    total += math.factorial(int(d))
  return total

def sumfacd(n):
  return sum([math.factorial(int(d)) for d in str(n)])

# all the code needed is in the main function
def main():
  chain60 = 0
  for n in range(1000000):
    next = n
    elems = set([next]) # set or list does not speed up the process
    l = 1
    while True:
      next = sum([math.factorial(int(d)) for d in str(next)])
      if next in elems:
        break
      else:
        elems.add(next)
      l += 1
      if l > 60:
        break
    if l == 60:
      chain60 += 1
  return chain60

main()
```
