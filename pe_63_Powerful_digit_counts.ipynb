```python
# Problem 63: Powerful digit counts

import math

count =0

for i in range(1, 23):
  start = math.ceil((10**(i-1))**(1/i))
  print(start)
  for j in range(start, 10):
    count += 1
    power = j**i
    print('%d - %d = %d^%d, %d digits' % (count, power, j, i, len(str(power))))
```

```python
# Problem 63: Powerful digit counts
k = 3
for i in range(10**(k-1),10**k):
  for j in range(2, 10):
    if i == j**k:
      print('%d = %d^%d' % (i, j, k))
```
