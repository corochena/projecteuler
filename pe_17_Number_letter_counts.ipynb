```python
units = ['', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine', 'ten', 'eleven', 'twelve', 'thirteen', 'fourteen', 'fifteen', 'sixteen', 'seventeen', 'eighteen', 'nineteen']
tens = ['zero', 'ten', 'twenty', 'thirty', 'forty', 'fifty', 'sixty', 'seventy', 'eighty', 'ninety']

def numToWords(n):
  if n < 20:
    return units[n]
  elif n < 100:
    d = int(n/10)
    u = n - d*10
    return tens[d] + numToWords(u)
  elif n < 1000:
    c = int(n/100)
    r = n - c*100
    if r > 0:
      mid = ' and '
    else:
      mid = ''
    return units[c] + ' hundred' + mid + numToWords(r)
  else:
    return "one thousand"

total = 0

for i in range(1,1001):
  s = numToWords(i).replace(' ', '')
  total += len(s)

print(total)
```
