```python
import urllib.request as url

minrange = list(range(ord('A'), ord('Z') + 1))
mayrange = list(range(ord('a'), ord('z') + 1))

pwd = ['e', 'x', 'p']

encrypted = url.urlopen('https://projecteuler.net/project/resources/p059_cipher.txt')

for line in encrypted:
  codes = line.decode('utf8').split(',')

def test():
  i = 2
 
  for val in range(ord('a'), ord('z') + 1):
    k = 0
    success = 0
    fail = 0
    while i + 3*k < len(codes):
      xor = val ^ int(codes[i + 3*k])
      if xor in minrange or xor in mayrange:
        success += 1
      else:
        fail += 1
      #print(xor, xor in minrange or xor in mayrange, chr(val), codes[i+3*k], i+3*k)
      k += 1
    print(chr(val), success, fail, success / (success + fail))

def decrypt():
  msg = ""
  sum = 0
  for i, code in enumerate(codes):
    xor = ord(pwd[i % 3]) ^ int(codes[i])
    sum += xor
    msg += chr(xor)
  print(sum, msg)

# test()

decrypt()
# 107 ^ 42

#for val in range(minrange[1], mayrange[0]+1):
#  print(chr(val))
```
