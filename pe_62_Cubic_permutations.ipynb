Sabemos que los cubos estan relativamente espaciados asi que primero hago un set con los cubos y reviso si entre ellos hay 5 anagramas (permutaciones). Esto lo hice al tanteo, empece con cifras de 8 digitos, luego 9, 10, 11 y en 12 encontre la respuesta.

```python
cubes = set()
xcube = 100000000000
x = int(xcube**(1/3)) + 1
while xcube <= 1e12:
  cubes.add(xcube)
  xcube = x**3
  x += 1
print(cubes)
```

```python
def isanagram(word1, word2):
  set1 = {l for l in word1}
  set2 = {l for l in word2}
  if set1 != set2:
    return False
  else:
    list2 = [c for c in word2]
    for l in word1:
      if l in list2:
        list2.remove(l)
      else:
        return False
    return True
```

```python
permutations = set()

while True:
  n1 = cubes.pop()
  permutations.add(n1)
  for n2 in cubes:
    if isanagram(str(n1), str(n2)):
      permutations.add(n2)
  if len(permutations) == 5:
    break
  else:
    permutations.clear()
  if len(cubes) < 5:
    break

print(permutations)

```
