```python
import math

def sumdigits(s):
  return sum([int(digit) for digit in s])

def econtfrac():
  euler = [2, 1]
  par = 2
  k = 0
  for i in range(98):
    if i % 3 == 0:
      euler.append(par + 2*k)
      k += 1
    else:
      euler.append(1)
  euler.reverse()
  #print(euler)

  n = 0
  d = 1

  for elem in euler:
    temp = n
    n = d
    d = elem * d + temp
  print(sumdigits(str(d)))

econtfrac()
```
