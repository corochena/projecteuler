```python
n = 1000000

mindiff = 1
winners = []

for d in range(1, n+1):
  if d % 7 != 0: # si es multiplo de 7 la fraccion es exactamente 3/7
    n = 3*d // 7 
    diff = 3/7 - n/d
    if diff < mindiff:
      mindiff = diff
      winners.append((n,d))

print(mindiff)
print(len(winners))
print(winners[-1])
```
