```python
import urllib.request

for line in urllib.request.urlopen('https://projecteuler.net/project/resources/p098_words.txt'):
  words = line.decode('utf-8').split(',')
  words = set(map(lambda s: s.replace('"',''), words))

dictwords = {}
for w in words:
  l = len(w)
  if l in dictwords.keys():
    dictwords[l].append(w)
  else:
    dictwords[l] = [w]
```

```python
for k,v in dictwords.items():
  print(k, len(v), v, len(v),k)
```

```python
def isanagram(word1, word2):
  set1 = {l for l in word1}
  set2 = {l for l in word2}
  if set1 != set2:
    return False
  else:
    list2 = [c for c in word2]
    for l in word1:
      if l in list2:
        list2.remove(l)
      else:
        return False
    return True

isanagram('RACAE','ACREE') 
```

```python
anagrams = []
for klen in dictwords:
  samelenlist = dictwords[klen][:]
  samelenset = set(samelenlist)
  for wordset in samelenset:
    samelenlist.remove(wordset)
    for wordlist in samelenlist:
      if isanagram(wordset, wordlist):
        anagrams.append((wordset, wordlist))
print(anagrams)
len(anagrams)
```

```python
import math

def anasq(word1, word2):
  if word1 == 'RACE':
    print(word1,word2)
  n = len(word1)
  letterset = {l for l in word1}
  squares = [x**2 for x in range(math.ceil((10**(n-1))**0.5), int((10**n-1)**0.5)+1)]
  for sq in squares:
    digits = {d for d in str(sq)}
    if len(digits) != len(letterset):
      continue
    equiv = {}
    for i,l in enumerate(word1):
      digit = str(sq)[i]
      if l in equiv:
        if equiv[l] != digit :
          break
      else:
        equiv[l] = str(sq)[i]
    if len(equiv) != n:
      continue
    num2 = ''
    for i in range(n):
      num2 += equiv[word2[i]]
    num2 = int(num2)
    if num2 in squares:
      print(equiv)
      return max(sq, num2)
  return 0

anasq('reduction', 'introduce')
```

```python
maxsq = 0
for pair in anagrams:
  w1, w2 = pair
  next = anasq(w1, w2)
  if next > maxsq:
    maxpair = w1, w2
    maxsq = next
print(maxsq, maxpair)
```
