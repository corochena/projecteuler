```python
def sumDigits(strnum):
  sum = 0
  for digit in strnum:
    sum += int(digit)
  return sum

def prob16():
  strnum = str(2**1000)
  print(sumDigits(strnum))

def prob18():
  fact100 = 1
  for i in range(2,101):
    fact100 *= i
  print(sumDigits(str(fact100)))

prob16()
```
