```python
# no se como hice para que funcionara de manera recursiva y combinando
# listas, cadenas y list comprehensions, YEAHH!!!
def allcombo(s):
  if len(s) == 1:
    return [s]
  else:
    elems = sorted([e for e in s])
    combo = []
    for e in elems:
      next = elems[:]
      next.remove(e)
      combo += prefix(e, allcombo(''.join(next)))
    return combo

def prefix(pf, l):
  return [pf + elem for elem in l]

lexi = allcombo('0123456789')
print(lexi[999999])
```
