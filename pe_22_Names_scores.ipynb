```python
import urllib.request as url

infile =url.urlopen('https://projecteuler.net/project/resources/p022_names.txt')
alpha = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

for line in infile:
  names = line.decode('utf-8').split(",")

names = list(map(lambda name: name[1:-1], names))
names.sort()

def nameVal(name):
  val = 0
  for letter in name:
    val += alpha.index(letter) + 1
  return val

total = 0
for pos, name in enumerate(names):
  total += (pos + 1)*nameVal(name)

print(total)
```
