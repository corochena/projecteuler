```python
import urllib.request as url
import numpy as np

file = url.urlopen('https://projecteuler.net/project/resources/p102_triangles.txt')

origins = 0

for line in file:
  coords = line.decode('utf8').split(',')
  vec1 = np.array((int(coords[0]), int(coords[1])))
  vec2 = np.array((int(coords[2]), int(coords[3])))
  vec3 = np.array((int(coords[4]), int(coords[5][0:-1])))
  ang1 = np.arccos(np.dot(vec1, vec2) / np.linalg.norm(vec1) / np.linalg.norm(vec2))
  ang2 = np.arccos(np.dot(vec1, vec3) / np.linalg.norm(vec1) / np.linalg.norm(vec3))
  ang3 = np.arccos(np.dot(vec3, vec2) / np.linalg.norm(vec3) / np.linalg.norm(vec2))
  total = abs(np.degrees(ang1)) + abs(np.degrees(ang2)) + abs(np.degrees(ang3))
  if abs(total - 360) < 0.001:
    origins += 1
  
print(origins)
```
