```python
def primesUpto(n):
  primes = set(range(2, n+1))
  multiples = set()

  for p in primes:
    i = 2
    while p*i <= n:
      multiples.add(p*i)
      i += 1
    if p > n**0.5:
      break
  return primes.difference(multiples)

print(sum(primesUpto(2000000)))

```
