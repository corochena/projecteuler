```python
def sumdigits(s):
  return sum([int(digit) for digit in s])
  
def mysecroot(x,n):
  introot = int(x**0.5)
  intx = x * 100
  for j in range(1, n + 1):
    introot = 10*introot
    for i in range(10):
      while introot * introot < intx:
        introot += 1
      introot -= 1
    intx *= 100
  return introot

def irratdigits():
  naturals = set(range(1,101))
  squares = set([1,4,9,16,25,36,49,64,81,100])
  nonperfect = naturals.difference(squares)
  
  sum = 0
  for x in nonperfect:
    sqrtofx = str(mysecroot(x, 99))
    sum += sumdigits(sqrtofx)
  print(sum)

irratdigits()
```
