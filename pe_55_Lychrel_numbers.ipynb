```python
def main():
    counter = 0
    for i in range(1,10000):
        if lychrel(i):
            counter += 1
            print('lychrel',i)
    print(counter)

def palindrome(s):
    for i in range(int(len(s)/2)):
        if s[i] != s[-i-1]:
            return False
    return True

def reverse(n):
    strn = str(n)
    rev = ''
    for c in strn:
        rev = c + rev
    return int(rev)

def lychrel(n):
    next = n
    for i in range(50):
        next += reverse(next)
        if palindrome(str(next)):
            #print(i, next)
            return False
    return True

main()
```
