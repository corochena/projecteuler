```python
def pentaList(size):
    return [int(0.5*n*(3*n - 1)) for n in range(1, size+1)]

def isPenta(p):
    root = (1 + (1 + 24*p)**0.5) / 6
    return int(root) == root

def indexPenta(p):
    return int((1 + (1 + 24*p)**0.5) / 6)

def penta(n):
    return int(0.5*n*(3*n-1))

n = 2170
penta_n = pentaList(n)

for k in range(n):
    for j in range(k+1, n):
        psum =  penta_n[k] + penta_n[j]
        pdif =  penta_n[j] - penta_n[k]
        if isPenta(psum) and isPenta(pdif):
            print(k + 1, penta_n[k], j + 1, penta_n[j], psum, pdif)

k=1020
j=2167
pk = penta(k)
pj = penta(j)
psum = pk + pj
pdif = pj - pk

print(pk, pj, psum, pdif)
print(isPenta(psum), indexPenta(psum),isPenta(pdif), indexPenta(pdif))
```
