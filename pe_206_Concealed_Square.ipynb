```python
start = int(1.02e18 ** 0.5)-3
end = int(1.93e18 ** 0.5)

print(start, end)

for x in range(start, end,10): # como el cuadrado termina en cero la base x debe ser multiplo de 10
  s = str(int(x**2))
  sub = s[2] + s[4] + s[6] + s[8] + s[10] + s[12] + s[14] + s[16] + s[18]
  if sub == "234567890":
    print(x, x**2, s)
```
