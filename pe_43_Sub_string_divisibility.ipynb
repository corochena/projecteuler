```python
# no se como hice para que funcionara de manera recursiva y combinando
# listas, cadenas y list comprehensions, YEAHH!!!
def allcombo(s):
  if len(s) == 1:
    return [s]
  else:
    elems = sorted([e for e in s])
    combo = []
    for e in elems:
      next = elems[:]
      next.remove(e)
      combo += prefix(e, allcombo(''.join(next)))
    return combo

def prefix(pf, l):
  return [pf + elem for elem in l]

lexi = allcombo('0123456789')
print(lexi[999999])
```

```python
def test(s):
  divlist = [2,3,5,7,11,13,17]
  start = 1
  for div in divlist:
    if int(s[start:start+3]) % div != 0:
      return False
    start += 1
  return True

print(test('1406357289'))

pandigsum = 0
for s in lexi:
  if test(s):
    print('found',s)
    pandigsum += int(s)
print(pandigsum)
```
