```python
# hacer una lista de los abundants
# del 24 al 28123 verificar cuales pueden ser escritos como la suma de dos abundants
# sumar todos los numeros que no pueden ser escritos como la suma de dos abundants

def abundants():
  abunList = []
  for i in range(1, 28123):
    if sum(setProperDivs(i)) > i:
      abunList.append(i)
  return abunList

def setProperDivs(n):
  orig = n
  divset = set([1])
  i = 2
  while n != 1:
    if n % i == 0:
      divls = list(divset)
      for elem in divls:
        divset.add(i*elem)
      divset.add(i)
      n /= i
    else:
      i += 1  
  divset.remove(orig)
  return divset

abunlist = abundants()
# usar un set en lugar de una lista signific una enorme diferencia en la rapidez
abunset = set(abunlist)
```

```python
# fue buena idea hacer el codigo en dos celdas, porque una vez teniendo la lista
# de los abundantes podia solo ejecutar la siguiente parte del codigo y ahorrarme
# 11 segundos cada vez que hiciera una prueba

def nonabunsum():
  impsum = 0
  for i in range(1, 28124):
    # aca uso una lista porque necesito seguir el orden de menor a mayor
    for a in abunlist:
      b = i - a
      if b < 0:
        impsum += i
        break
      # aca uso un set para buscar mas rapido
      if b in abunset:
        break
  return impsum

nonabunsum()
```
