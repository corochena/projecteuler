```python
# obtengo los primos menores que 1 millon (o 10 millones talvez)
# obtengo los compuestos menores que 1 millon
# obtengo los compuestos impares menores que 1 millon
# para cada compuesto impar intento escribirlo en la forma primo + 2*x^2 donde x es un natural
# si no puedo escribirlo ya obtuve mi respuesta

def primesOrig(n):
  primes = [2]

  for num in range(primes[-1] + 1, n + 1):
    isPrime = True
    for div in primes:
      if num % div == 0:
        isPrime = False
        break
      if div > num**0.5:
        break
    if isPrime:
      primes.append(num)

  return primes

shortlist = primesOrig(10000)
shortset = set(shortlist)
oddcomp = []
for i in range(9,10000,2):
  if not i in shortset:
    oddcomp.append(i)
print(len(oddcomp), oddcomp[:20])
```

```python
def goldbach3():
  for oc in oddcomp:
    for p in shortlist:
      if p > oc:
        return oc
      i = ((oc - p)/2)** 0.5
      if i == int(i):
        #print('oc', oc, '=', p, '+ 2(', int(i), ')^2')
        break
       
goldbach3()
```

```python
def goldbach2():
  for oc in oddcomp:
    found = True
    for p in shortlist:
      if not found:
        break
      #print(oc, p)
      if p > oc and found:
        return oc
      else:
        found = True
        i = ((oc - p) / 2) ** 0.5
        if i == int(i):
          found = False
       
goldbach2()
```

```python
def goldbach():
  for oc in oddcomp:
    found = True
    for p in shortlist:
      if not found:
        break
      #print(oc, p)
      if p > oc and found:
        return oc
      else:
        found = True
        i = 1
        x = p + 2*i**2
        if x == oc:
          #print(i, p, oc, x, found)
          found = False
          break
        while x < oc:
          i += 1
          x += 4*i - 2
          if x == oc:
            found = False
            #print(i, p, oc, x, found)
            break
        

goldbach()
```
